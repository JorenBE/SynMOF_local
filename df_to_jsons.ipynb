```python
import pandas as pd
import os
import json


FILE = '/home/jorenvanherck/Documents/synmofWeb/local_loop/20250508_1508_predictions_all.csv'
SAVE_FOLDER = '/home/jorenvanherck/Documents/synmofWeb/local_loop/predictions_json'
```

```python

df = pd.read_csv(
    FILE,
    index_col=0
)

# 2) Convert each row to a dict
records = df.to_dict(orient="records")

print(records)
```

```python
for i, rec in enumerate(records):
    # e.g. json_rows/row_0.json, row_1.json, …
    fname = os.path.join(SAVE_FOLDER, f"{rec['name']}.json")
    with open(fname, "w", encoding="utf-8") as f:
        json.dump(rec, f, ensure_ascii=False, indent=2)

print(f"Wrote {len(records)} JSON files to {SAVE_FOLDER}")
```
